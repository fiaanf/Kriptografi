```python
def vigenere_encrypt(plain, key):
    res = ''
    for i in range(len(plain)):
        res += chr(((ord(plain[i]) - 65 + ord(key[i % len(key)]) - 65) % 26) + 65)
    return res
print(vigenere_encrypt('CRYPTOGRAPHY','LEMON'))
```

```python
 def vigenere_decrypt(cipher, key):
     res = ''
     for i in range(len(cipher)):
         res += chr(((ord(cipher[i]) - ord(key[i % len(key)])) % 26) + 65)
     return res
 print(vigenere_decrypt('NFFXZTTBDZNL','LEMON'))
 print(vigenere_decrypt('LXFOPVEFRNHR','LEMON'))
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Teks yang akan dianalisis
text = 'NFFXZITB D ZNL'.replace(' ', '')  # hilangkan spasi biar bersih
count = Counter(text)

# Hitung frekuensi relatif
letters = list(count.keys())
frequencies = [v / len(text) for v in count.values()]

# Visualisasi frekuensi
plt.figure(figsize=(8, 5))
plt.bar(letters, frequencies)
plt.title('Frekuensi Huruf pada Ciphertext (Vigenere Cipher)')
plt.xlabel('Huruf')
plt.ylabel('Frekuensi Relatif')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
